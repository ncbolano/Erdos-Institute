```python
def craps_pass_line():
    come_out = np.random.randint(1,7) + np.random.randint(1,7)
    if come_out in [7,11]:
        return 1
    elif come_out in [2,3,12]:
        return 0
    else: 
        point = come_out 

        while True:
            roll = np.random.randint(1,7) + np.random.randint(1,7)
            if roll == 7:
                return 0 
            elif roll == point:
                return 1

craps_games = [craps_pass_line() for _ in range(1000)]
win_rate = np.mean(craps_games)
print(f"Percentage of wins in 1000 games: {win_rate * 100}%")
```
